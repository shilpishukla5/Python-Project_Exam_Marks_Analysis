```python
ax=sns.countplot(x= 'Gender',data=df, color='green')
ax.bar_label(ax.containers[0])
plt.figure(figsize=(1,5))

```

```python
gb_PE=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print( gb_PE)
```

```python
sns.heatmap(gb_PE,annot=True,).set(title='Relationship between parents education and students marks')

#plt.title("Relationship between Parent Education and student marks")
plt.figure(figsize=(5,5))

plt.show()
```

With this heat map we can observe that there is significant impact on student's marks with parents having master degree and having high school degree

```python
gb_MS=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print( gb_MS)
```

```python
sns.heatmap(gb_MS,annot=True).set(title='Relationship between parents marital status and students marks')
plt.show()

```

With this heat map we can conclude that there is no major effect of parents marital status on their children's marks
